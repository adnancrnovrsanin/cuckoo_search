```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the CSV file
csv_file = "cuckoo_search_results.csv"
df = pd.read_csv(csv_file)

# Filter for Unimodal benchmark functions
unimodal_functions = ["sphere", "sumSquares", "step2", "quartic", "rosenbrock", "schwefel1_2", "schwefel2_21", "schwefel2_22"]
df_unimodal = df[df["Benchmark Function"].isin(unimodal_functions)]

# Set the style for the plots
sns.set(style="whitegrid")

# Visualization of unimodal benchmark functions
plt.figure(figsize=(14, 8))
sns.lineplot(data=df_unimodal, x="Number of Threads", y="Min Time", hue="Benchmark Function", marker="o", linestyle="-")
sns.lineplot(data=df_unimodal, x="Number of Threads", y="Avg Time", hue="Benchmark Function", marker="o", linestyle="--")
sns.lineplot(data=df_unimodal, x="Number of Threads", y="Max Time", hue="Benchmark Function", marker="o", linestyle=":")

# Adding custom legend for line types
from matplotlib.lines import Line2D

legend_elements = [
    Line2D([0], [0], color='k', linestyle='-', label='Min Time'),
    Line2D([0], [0], color='k', linestyle='--', label='Avg Time'),
    Line2D([0], [0], color='k', linestyle=':', label='Max Time')
]

# Adding custom legend for colors
handles, labels = plt.gca().get_legend_handles_labels()
unique_labels = list(dict.fromkeys(labels))  # Remove duplicates
unique_handles = handles[:len(unique_labels)]  # Keep only unique handles

plt.legend(handles=unique_handles + legend_elements, title='Benchmark Function and Time Type', loc='upper center', bbox_to_anchor=(0.5, -0.1), ncol=2)

# Set titles and labels
plt.title('Performance of Cuckoo Search Algorithm on Different Number of Threads (Unimodal Functions)')
plt.xlabel('Number of Threads')
plt.ylabel('Execution Time (seconds)')

# Display the plot
plt.tight_layout()
plt.show()

# Separate plots for min, avg, and max execution times
fig, axes = plt.subplots(3, 1, figsize=(14, 24), sharex=True)

# Min Time
sns.lineplot(ax=axes[0], data=df_unimodal, x="Number of Threads", y="Min Time", hue="Benchmark Function", marker="o", linestyle="-")
axes[0].set_title('Min Execution Time (Unimodal Functions)')
axes[0].set_ylabel('Min Time (seconds)')
axes[0].legend(loc='upper center', bbox_to_anchor=(0.5, -0.1), ncol=2)

# Avg Time
sns.lineplot(ax=axes[1], data=df_unimodal, x="Number of Threads", y="Avg Time", hue="Benchmark Function", marker="o", linestyle="--")
axes[1].set_title('Avg Execution Time (Unimodal Functions)')
axes[1].set_ylabel('Avg Time (seconds)')
axes[1].legend(loc='upper center', bbox_to_anchor=(0.5, -0.1), ncol=2)

# Max Time
sns.lineplot(ax=axes[2], data=df_unimodal, x="Number of Threads", y="Max Time", hue="Benchmark Function", marker="o", linestyle=":")
axes[2].set_title('Max Execution Time (Unimodal Functions)')
axes[2].set_xlabel('Number of Threads')
axes[2].set_ylabel('Max Time (seconds)')
axes[2].legend(loc='upper center', bbox_to_anchor=(0.5, -0.1), ncol=2)

# Display the plots
plt.tight_layout()
plt.show()
```
